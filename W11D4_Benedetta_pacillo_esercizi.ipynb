# Esercizio 2 

Creiamo un DataFrame sintetico, che contiene i guadagni mensili di diverse annate, con il seguente codice: years = 5 a = pd.DataFrame({"Mese": list("GFMAMGLASOND"*years), "Anno": np.repeat(list(range(years)), 12), "Guadagni": np.random.randint(800, 5000, 12*years)}) • Calcola la somma cumulativa delle vendite per ogni mese utilizzando il metodo .cumsum() • Come sopra, ma raggruppato per ogni anno usando prima un .groupby()

```python
years = 5 
a = pd.DataFrame({"Mese": list("GFMAMGLASOND"*years), "Anno": np.repeat(list(range(years)), 12), "Guadagni": np.random.randint(800, 5000, 12*years)})
```

```python
a.cumsum()
```

```python
a['year_cumsum'] = a.groupby('Anno').Guadagni.cumsum()
a
```

Estraiamo solo i post relativi al 2012 • Estraiamo solo i post relativi a maggio 2018 • Confrontiamo il numero di post pubblicati nei weekend rispetto al numero di post pubblicati nel resto della settimana • Troviamo il primo e ultimo post pubblicati in ogni anno • Quanti tipi di post ci sono? E quanti per ogni tipo?

```python
flt_2012 = fb.status_published.dt.year == 2012
fb.status_published[flt_2012]
```

```python
flt_2018 = fb.status_published.dt.year == 2018
flt_05 = fb.status_published.dt.month == 5

fb.status_published[flt_2018 & flt_05]
```

```python
weekday_post = fb.status_published.dt.dayofweek < 5
weekend_post = fb.status_published.dt.dayofweek >= 5

print('numero di post nel weekend: ', fb[ weekend_post ].shape[0])
print('numero di post negli altri giorni: ', fb[ weekday_post ].shape[0])
```

```python
fb_year = fb.status_published.dt.year

for year in fb_year.unique():
    fb_year_post = fb.groupby(fb_year).get_group(year).sort_values('status_published')
    print(year)
    display('Primo post:', fb_year_post.head(1))
    display('Ultimo post: ', fb_year_post.tail(1))
```
